```python
all_files=[]
import os
for root, dirs, files in os.walk("../valve1_pred/"):
    for file in files:
        if file.endswith(".csv"):
             all_files.append(os.path.join(root, file))
```

```python
# datasets with anomalies loading
list_of_df = [pd.read_csv(file, 
                          sep=';', 
                          index_col='datetime', 
                          parse_dates=True) for file in all_files]
```

```python
# inference
def func(df,Q = 0.9):
    realDF = df.iloc[:,0:8]
    
    scaler = StandardScaler()
    train_data = realDF.iloc[0:400,:]
    scaler.fit(train_data.values)
    real_scaled=scaler.transform(realDF) 
    real_scaledDF = pd.DataFrame(real_scaled)
    
    predDF = df.iloc[:,10:]
    predDF.reset_index(drop=True,inplace=True)
    
    realLabel = df.iloc[:,8:10]
    realLabel.reset_index(drop=True,inplace=True)
    
    con_DF = pd.concat([real_scaledDF,predDF],axis=1)
    
    # results predicting
    X = np.array(real_scaledDF)
    PredictX = np.array(predDF)
    informer_residuals = pd.Series(np.mean(np.abs(X - PredictX), axis=1))
    
    # quantile for upper control limit (UCL) selection
    UCL = informer_residuals.quantile(Q)
    
    # data i is an anomaly if samples [(i - timesteps + 1) to (i)] are anomalies
    anomalous_data = informer_residuals > UCL
    
    anomalous_data_indices = []
    N_STEPS = 20
    for data_idx in range(N_STEPS - 1, len(X) - N_STEPS + 1):
        if np.any(anomalous_data[data_idx - N_STEPS + 1 : data_idx]):
            anomalous_data_indices.append(data_idx)
    
    prediction = pd.Series(data=0, index=df.index)
    prediction.iloc[anomalous_data_indices] = 1
    
    # true outlier indices selection
    true_outlier = df.anomaly 
    return true_outlier, prediction
```

```python
# binary classification metrics calculation
def evaluating_change_point(true_outlier, predicted_outlier):
    Nsum = len(true_outlier)
    rights = (true_outlier == predicted_outlier)
    print(rights)
    accuracy = np.sum(rights)/Nsum
    TP = np.sum((true_outlier==0) & (predicted_outlier==0))
    FN = np.sum((true_outlier==0) & (predicted_outlier==1))
    FP = np.sum((true_outlier==1) & (predicted_outlier==0))
    TN = np.sum((true_outlier==1) & (predicted_outlier==1))
    print(TP,FN,FP,TN)
    P = TP/(TP+FP)
    R = TP/(TP+FN)
    FAR = FP/(FP+TN)
    MAR = FN/(TP+FN)
    F1 = 2*P*R/(P+R)
    return accuracy,P,R,F1,FAR,MAR
```

```python
true_outliers = []
predicted_outliers = []
for df in list_of_df:
    true_outlier, prediction = func(df,0.9)
    true_outlier = true_outlier.tolist()
    true_outlier=list(map(int,true_outlier))
    prediction = prediction.tolist()
    true_outliers.extend(true_outlier)
    predicted_outliers.extend(prediction)
```
